```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[50,40,30,20,35,45,55]
```

```python
plt.plot(x,y,'bo')
plt.plot(x,y)
```

```python
plt.plot(x,y,'go--', linewidth=2, markersize=12)
```

```python
#I will give us marker in diamond shape
plt.plot(x, y, color='green', marker='D', linestyle='',
linewidth=2, markersize=5)
```

```python
#Alpha(Control the transperency of the line between 0 to 1)
plt.plot(x,y,color='Green',alpha=0.5)
```

```python
#Axis ,Label,Lengend ,Grid
```

```python
days=[1,2,3,4,5,6,7]
max_t=[51,52,53,54,55,56,57]
min_t=[31,32,33,34,35,36,37]
avg_t=[41,42,43,44,45,46,47]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
#To give name to axis and give title to graph
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
```

```python
#Label and #Grid
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend(loc='best')
plt.grid()#It will show the grid in background
```

# scatter plot

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
x,y
```

```python
plt.scatter(x,y,color='g',label='Points')
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title("scatter")
```
